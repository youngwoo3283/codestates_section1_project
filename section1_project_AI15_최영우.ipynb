## sale 값 조정하기
 - k나 m을 수치로 변경하기
 - object을 int로 변경

```python
df.columns
```

```python
#얘는 sale가 M이나 K가 붙어있으니 이를 위처럼 apply로 다시 조정한다.
#원래 million단위이니 M은 빼고 K는 1000을 의미하니 0.001을 곱해준다.
df.loc[11]
```

```python
def change(x):
  if x[-1] == 'K':
    x = x.replace('K','')
    x  = float(x) * 0.001
    return x
  elif x[-1] == 'M':
    x =  x.replace('M','')
    x = float(x)
    return x
  else:
    x = float(x)
    return x
```

```python
df['NA_Sales'] = df['NA_Sales'].apply(func = change)
```

```python
df.loc[11] #NA_Sales가 숫자로 바꿔는것을 확인
```

```python
df['JP_Sales'] = df['JP_Sales'].apply(func = change)
df['EU_Sales'] = df['EU_Sales'].apply(func = change)
df['Other_Sales'] = df['Other_Sales'].apply(func = change)
```

```python
df.loc[11] #나머지가 숫자로 바꿔는것을 확인하였다.
```

```python
#모두 flaot으로 바꿔줌
df['NA_Sales'].astype(float)
df['EU_Sales'].astype(float)
df['JP_Sales'].astype(float)
df['Other_Sales'].astype(float)
```

```python
#판매 총합 칼럼만들기
df['sum_sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']
```

# 결론
- 기존의 게임 데이터를 보니 액션과 스포츠 장르가 꾸준한 인기를 얻는 게임 장르이다.
- 지역별 장르의 출고량을 보니 미국과 유럽시장의 경우는 액션과 스포츠 및 fps장르를 출시하면 되지만 일본에 출시할 경우에는 fps는 인기가 없고 롤플레잉이 제일 인기가 많다는 것을 유의해야 한다.
- 전체 게임시장에 대한 지역의 출고량 비율을 살펴보니 유럽과 미국의 시장출고량이 75%정도이니 일본시장을 꼭 겨냥해서 롤플레잉 장르를 할 필요는 없다고 생각된다.
- 2000년 전에는 puzzle과 platform racing등의 게임이 인기가 있었다. 
- 그 뒤로는 action과 sport shooter가 인기가 꾸준히 있어서 현재 트렌드는 **action, sport, shooter**장르이다
- 판매량이 높은 10개의 게임은 모두 **닌텐도**가 배급을 하였다.
-  닌텐도와 EA 액티비젼이 출고량이 높은데 EA와 액티비젼의 경우 일본에 출고량이 낮다. 따라서 **닌텐도**와 퍼블리싱을 맺고 게임을 출시하는 것이 제일 이상적이다.
- 메타스코어 데이터를 보니 스포츠의 경우 제일 높은 평균 평점을 기록하였고 액션의 경우는 비교적 낮은 것을 볼 수 있다.
- 2010년이후로 데이터를 보니 **action**는 여전히 인기가 많지만 sports장르는 인기가 떨어지는 것을 확인 할 수 있다. **adventure**와 **role-playing**이 인기가 많은 것을 확인 할 수 있다.
-  pc 및 닌텐도사의 switch가 가장 많이 출시하는 것을 알 수 있다.





## 따라서 액션 장르가 꾸준히 시대별로 인기가 있고 전세계 출고 점유율도 가장 많으니 액션 장르의 게임을 출시하는 것이 이상적이다. 이때 액션 게임의 메타스코어가 낮은 편이니 게임의 퀄리티를 높이는 요소(스토리, 작화, 그래픽, 사운드)가 중요하다. 그리고 출시 플랫폼은 닌텐도 스위치 혹은 PC로 출시하는 것이 좋고 배급사의 경우 닌텐도 혹은 액티비젼을 선택하는 것이 이상적이라고 생각한다.
